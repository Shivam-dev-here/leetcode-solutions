# Find Numbers with Even Number of Digits

## Solutions

```python
# Not optimized

class Solution(object):
    def findNumbers(self, nums):
        counted=0
        for num in nums:
            count = 0
            while num!=0:
                count+=1
                num //= 10
            if count % 2 ==0:
                counted+=1
        return counted
```

```python
nums = [555,901,482,1771]
counter = []
for num in nums:
    count=0
    while num!=0:
        if num==0:
            break
        else:
            count+=1
            num //= 10
    counter.append(count)
nums = 0
for count in counter:
    if count % 2 ==0:
        nums+=1

print(nums)
```
