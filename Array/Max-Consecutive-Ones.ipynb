```python
class Solution(object):
    def findMaxConsecutiveOnes(self, nums):
        ci=1
        co=1
        for i in range(0,len(nums)):
            if len(nums) == i+2:
                break
            elif(nums[i]== 1 and nums[i]==nums[i+1]):
                ci+=1
            elif(nums[i]!=nums[i+1] and co<ci):
                co = ci
                ci=1
        return co

nums = [1,1,0,1,1,1]
print(Solution().findMaxConsecutiveOnes(nums))
```
