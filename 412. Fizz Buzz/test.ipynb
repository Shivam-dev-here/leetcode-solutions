# Solution

```python
class Solution(object):
    def fizzBuzz(self, n):
        arr = []
        for i in range(1, n+1):
            div3 = (i % 3) == 0
            div5 = (i % 5) == 0
            if (div3 and div5) == 1:
                arr.append("FizzBuzz")       
            elif div3 == 1:
                arr.append("Fizz")
            elif div5 ==1:
                arr.append("Buzz")
            else:
                arr.append(str(i))
        return arr
```
